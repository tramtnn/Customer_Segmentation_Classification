```python
import datetime 
current_day = max(bill['create_info.created']) + datetime.timedelta(days=1)

df_customers = bill.groupby(['customer_id']).agg(
    {'create_info.created' : lambda x: (current_day -x.max()).days,
     '_id': 'count',
     ' summary.gran _total ' :'sum'    
    }
)
df_customers.rename(columns={'create_info.created':'R', '_id':'F', ' summary.gran _total ':'M'}, inplace =True)
```

```python
df_customers
```

```python
fig, ax = plt.subplots( 1, 3, figsize=(12,4))
sns.distplot(df_customers['R'], ax=ax[0])
sns.distplot(df_customers['F'], ax=ax[1])
sns.distplot(df_customers['M'], ax=ax[2])
```

```python
from scipy import stats
def analyze_skewness(x):
    fig, ax = plt.subplots(3, 1, figsize=(12,7))
    sns.distplot(df_customers[x], ax=ax[0])
    sns.distplot(np.log(df_customers[x]), ax=ax[1])
    sns.distplot(np.sqrt(df_customers[x]), ax=ax[2])
    plt.tight_layout()
    plt.show()
    
    print(df_customers[x].skew().round(2))
    print(np.log(df_customers[x]).skew().round(2))
    print(np.sqrt(df_customers[x]).skew().round(2))
```

```python
analyze_skewness('F')
```

```python
df_customers_log = np.log(df_customers)
fig, ax = plt.subplots( 1, 3, figsize=(12,4))
sns.distplot(df_customers_log['R'], ax=ax[0])
sns.distplot(df_customers_log['F'], ax=ax[1])
sns.distplot(df_customers_log['M'], ax=ax[2])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df_customers_log)

df_customers_t = scaler.transform(df_customers_log)
df_customers_t
```

```python
df_customers_t= pd.DataFrame(df_customers_t)
```

```python
df_customers_t
```

```python
from sklearn.cluster import KMeans
df_customers_t = df_customers_t.fillna(df_customers_t.mean())

sse = {}
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_customers_t)
    sse[k] = kmeans.inertia_ 
    
plt.title('The Elbow Method')
plt.xlabel('k')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

Chọn K=4

```python
model = KMeans(n_clusters=6, random_state=42)
model.fit(df_customers_t)
```

```python
df_customers['Cluster']= model.labels_
df_customers.head(10)
```

```python
df_customers.groupby('Cluster').agg(
    {
        'R':'mean',
        'F':'mean',
        'M':'mean'
    }
).round(2)
```

```python
bill['Unnamed: 13'] = bill['to_user.date'].dt.month
bill.head()
```

```python
bill = bill.rename(columns={'Unnamed: 13':'month'})
monthly_totals = bill.groupby(['year', 'month'])[' summary.gran _total '].sum().reset_index()
monthly_totals.head()
```

```python
top5_months = monthly_totals.groupby('year').apply(lambda x: x.nlargest(5, ' summary.gran _total ')).reset_index(drop=True)
print(top5_months)
```

```python
for year, data in top5_months.groupby('year'):
    plt.figure(figsize=(10, 6))
    plt.bar(data['month'], data[' summary.gran _total '], label=str(year), color='pink')
    plt.title(f'Top 5 tháng có đơn hàng nhiều nhất năm {year}')
    plt.xlabel('Tháng')
    plt.ylabel('Tổng giá trị đơn hàng')
    plt.legend()
    plt.show()
```
